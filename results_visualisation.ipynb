```python
import numpy as np
import matplotlib.pyplot as plt

def compute_moving_average(data: np.ndarray, window_size: int) -> np.ndarray:
    """
    Compute moving average.
    Args:
        data (np.ndarray): data to compute moving average on
        window_size (int): window size
    Returns:
        mv_avg (np.ndarray): moving average of the data
    """
    cumulative_sum = np.cumsum(data)
    cumulative_sum[window_size:] = cumulative_sum[window_size:] - cumulative_sum[:-window_size]
    mv_avg = cumulative_sum[window_size - 1:] / window_size
    return mv_avg

# Load the scores
scores = np.load("results/scores_512_entropy_coef_0.005.npy")

# Compute moving average
moving_avg = compute_moving_average(scores, 100)

plt.figure(figsize=(10,5))
plt.plot(scores, alpha=0.6, label='Score per Episode')
plt.plot(range(99, len(scores)), moving_avg, color='red', label='Moving Average (over 100 episodes)')
plt.xlabel('Episode')
plt.ylabel('Score')
plt.title('Scores and Moving Average over 100 Episodes')
plt.legend()
plt.grid(True)
plt.show()
```
